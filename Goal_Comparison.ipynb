```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from PIL import Image

sns.set_style('darkgrid')
```

```python
df = pd.read_feather('../../Data Source/Football Transfer/4_seasons_goals_record')

df.head()
```

```python
df = df.sort_values('season')
```

```python
df['season']
```

```python
bigsix = ['Chelsea FC', 'Manchester City', 'Liverpool FC','Arsenal FC', 'Tottenham Hotspur', 'Manchester United' ]
```

```python
df[df['name'] == 'Chelsea FC'].iloc[0,2]
```

```python
bigsix = ['Chelsea FC', 'Manchester City', 'Liverpool FC','Arsenal FC', 'Tottenham Hotspur', 'Manchester United' ]

fig, axes = plt.subplots(figsize=(10,5))

for club in bigsix:
    club_data = df[df['name'] == club]
    sns.lineplot(data=club_data, x='season', y='Goal', label=club, ax=axes)
    sns.scatterplot(data=club_data, x='season', y='Goal', ax=axes)
    if club == 'Chelsea FC':
        mean_goal = club_data['Goal'].mean()
        for i in range(0, len(club_data['season'])):
            axes.text(
                club_data['season'].iloc[i],
                club_data['Goal'].iloc[i] + 2,
                club_data['Goal'].iloc[i],
                horizontalalignment='center', 
                size='medium', 
                color='black', 
                weight='semibold'
            )
        
axes.axhline(y=mean_goal, ls='--', color='#996633')
axes.text(3.15, mean_goal-1, f'Chelsea Goal Mean: {mean_goal:.0f}')

plt.ylim(0, 110)
plt.legend(loc='best', bbox_to_anchor = (1,1))  
plt.title('Big Six Goal Scoring Comparison in PL')  
plt.show()
```

```python
bigsix = ['Chelsea FC', 'Manchester City', 'Liverpool FC','Arsenal FC', 'Tottenham Hotspur', 'Manchester United' ]

fig, axes = plt.subplots(figsize=(10,5))

for club in bigsix:
    club_data = df[df['name'] == club]
    sns.lineplot(data=club_data, x='season', y='Loss', label=club, ax=axes)
    sns.scatterplot(data=club_data, x='season', y='Loss', ax=axes)
    if club == 'Chelsea FC':
        mean_loss = club_data['Loss'].mean()
        for i in range(0, len(club_data['season'])):
            axes.text(
                club_data['season'].iloc[i],
                club_data['Loss'].iloc[i] + 2,
                club_data['Loss'].iloc[i],
                horizontalalignment='left', 
                size='medium', 
                color='black', 
                weight='semibold'
            )
        
axes.axhline(y=mean_loss, ls='--', color='black')
axes.text(3.1, mean_loss+0.5, f'Chelsea Mean Loss: {mean_loss:.0f}')
    
plt.legend(loc='best', bbox_to_anchor = (1,1))  
plt.title('Big Six Goal Scoring Comparison in PL')  
plt.show()
```

```python
bigsix = ['Chelsea FC', 'Manchester City', 'Liverpool FC','Arsenal FC', 'Tottenham Hotspur', 'Manchester United' ]
df['net goal'] = df['Goal'] - df['Loss']

fig, axes = plt.subplots(figsize=(10,5))

for club in bigsix:
    club_data = df[df['name'] == club]
    sns.lineplot(data=club_data, x='season', y='net goal', label=club, ax=axes)
    sns.scatterplot(data=club_data, x='season', y='net goal', ax=axes)
    if club == 'Chelsea FC':
        mean_loss = club_data['net goal'].mean()
        for i in range(0, len(club_data['season'])):
            axes.text(
                club_data['season'].iloc[i],
                club_data['net goal'].iloc[i] + 2,
                club_data['net goal'].iloc[i],
                horizontalalignment='center', 
                size='medium', 
                color='black', 
                weight='semibold'
            )
        
axes.axhline(y=mean_loss, ls='--', color='black')
axes.text(3.1, mean_loss+0.5, f'Chelsea Mean net goal: {mean_loss:.0f}')
    
plt.legend(loc='best', bbox_to_anchor = (1,1))  
plt.title('Big Six Goal Scoring Comparison in PL')  
plt.show()
```
